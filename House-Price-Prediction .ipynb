```python
x=ab[[ 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
       'floors', 'waterfront', 'view', 'condition', 'sqft_above',
       'sqft_basement']]
x
```

```python
y=ab[['price']]
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn import preprocessing
lab_enc = preprocessing.LabelEncoder()
training_scores_encoded = lab_enc.fit_transform(ytrain)
clf = DecisionTreeClassifier()
clf.fit(xtrain, training_scores_encoded)

confidence = clf.score(xtrain, training_scores_encoded)
print(confidence)
```

```python
fig,ax=plt.subplots(1,1,figsize=(18,12))
sns.heatmap(ab.corr(),annot=True,linewidths=1)
plt.show()
```

```python
prediction=clf.predict([[5.0,2.50,3650,9050,2.0,0,4,5,3370,280]])
print(prediction)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn import preprocessing
lab_enc = preprocessing.LabelEncoder()
training_scores_encoded = lab_enc.fit_transform(ytrain)
clf =AdaBoostClassifier()
clf.fit(xtrain, training_scores_encoded)

confidence = clf.score(xtrain, training_scores_encoded)
print(confidence)
```

```python
prediction=clf.predict([[3.0,1.50,1340,7912,1.5,0,0,3,1340,0]])
print(prediction)
```
